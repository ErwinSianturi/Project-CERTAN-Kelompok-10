```python
import os
from tensorflow.keras import layers
from tensorflow.keras import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint, EarlyStopping
from tensorflow.keras.models import Sequential,load_model,save_model
from tensorflow.keras.layers import Dense,Conv2D,Flatten,MaxPooling2D
from keras.layers import BatchNormalization
from keras.optimizers import Adam
```

```python
import os
import shutil
from PIL import Image
from sklearn.model_selection import train_test_split

# Directories
dataset_dir = r"C:\\Users\\erwin\\Certan Last\\Testing\\dataset"
train_dir = r"C:\\Users\\erwin\\Certan Last\\Testing\\train"
val_dir = r"C:\\Users\\erwin\\Certan Last\\Testing\\validation"

# Create train and validation folders
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Categories (e.g., 'man', 'woman')
categories = ['man', 'woman']

# Minimum size requirement
min_width, min_height = 65, 65

# Process each category
for category in categories:
    category_path = os.path.join(dataset_dir, category)
    images = os.listdir(category_path)
    
    # Filter images by size
    valid_images = []
    for image_name in images:
        image_path = os.path.join(category_path, image_name)
        try:
            with Image.open(image_path) as img:
                width, height = img.size
                if width >= min_width and height >= min_height:
                    valid_images.append(image_name)
        except Exception as e:
            print(f"Error processing image {image_name}: {e}")
    
    # Split the valid images into train and validation sets
    train_images, val_images = train_test_split(valid_images, test_size=0.2, random_state=42)
    
    # Create category folders in train and validation directories
    train_category_path = os.path.join(train_dir, category)
    val_category_path = os.path.join(val_dir, category)
    os.makedirs(train_category_path, exist_ok=True)
    os.makedirs(val_category_path, exist_ok=True)
    
    # Copy train images
    for image in train_images:
        shutil.copy(os.path.join(category_path, image), os.path.join(train_category_path, image))
    
    # Copy validation images
    for image in val_images:
        shutil.copy(os.path.join(category_path, image), os.path.join(val_category_path, image))

print("Dataset has been filtered and divided into Train and Validation sets.")

```

```python
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,  # Rotasi gambar
    width_shift_range=0.2,  # Perpindahan horizontal
    height_shift_range=0.2,  # Perpindahan vertikal
    shear_range=0.2,  # Distorsi shearing
    zoom_range=0.2,  # Zoom in dan zoom out
    horizontal_flip=True,  # Membalik gambar secara horizontal
    fill_mode='nearest'  # Menangani area kosong akibat augmentasi
)

test_datagen = ImageDataGenerator(rescale=1.0/255)
```

```python
batch_size = 64
target_size = (64, 64)
input_shape=(64, 64, 3)
seed=1337
adam = 0.005
fre= -20
FC = 2048
E = 1
patience = 7
verbose = 1
factor = 0.50
min_lr = 0.001
steps_per_epoch=128
validation_steps=128
epochs=20
```

```python
train_generator = train_datagen.flow_from_directory(
    'C:\\Users\\erwin\\Certan Last\\Testing\\train',
    batch_size=batch_size,
    class_mode='binary',
    seed=seed,
    target_size=target_size
)

validation_generator = test_datagen.flow_from_directory(
    'C:\\Users\\erwin\\Certan Last\\Testing\\validation',
    batch_size=batch_size,
    class_mode='binary',
    seed=seed,
    target_size=target_size
)
```

```python
base_model = tf.keras.applications.VGG16(input_shape=input_shape,include_top=False,weights="imagenet")
```

```python
fre = len(base_model.layers)  # Freezing all layers
for layer in base_model.layers[:fre]:
    layer.trainable = False
```

```python
# Building Model
from keras.layers import BatchNormalization
from keras.regularizers import l2
model=Sequential()
model.add(base_model)
model.add(layers.Dropout(.2))

model.add(Conv2D(512, (3, 3),strides=(1,1), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(layers.Dropout(.1))
model.add(Conv2D(128, (3, 3),strides=(1,1), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(layers.Dropout(0.5))
model.add(Conv2D(384, (3, 3),strides=(1,1), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(layers.Dropout(.1))
model.add(Conv2D(384, (3, 3),strides=(1,1), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(layers.Dropout(.1))
model.add(Conv2D(500, (3, 3),strides=(1,1), activation='relu', padding='same'))
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,strides=(2,2), padding='same'))



# Add new layers
model.add(Flatten())
model.add(Dense(FC , activation='relu'))
model.add(layers.Dropout(.2))
model.add(Dense(FC , activation='relu'))
model.add(layers.Dropout(.2))
model.add(Dense(FC, activation='relu'))
model.add(layers.Dropout(.2))
model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(E, activation='sigmoid'))

model.summary()
```

```python
# Callback untuk pelatihan
from keras.callbacks import ReduceLROnPlateau, ModelCheckpoint, EarlyStopping

# Callback untuk menurunkan learning rate secara adaptif
lrd = ReduceLROnPlateau(
    monitor='val_loss',
    patience=patience,
    verbose=verbose,
    factor=factor,
    min_lr=min_lr
)

# Callback untuk menyimpan model terbaik selama pelatihan
mcp = ModelCheckpoint(
    'model.keras',  # Format terbaru untuk file model
    monitor='val_loss',
    save_best_only=True,
    verbose=verbose
)

# Callback untuk menghentikan pelatihan lebih awal jika tidak ada peningkatan
es = EarlyStopping(
    verbose=verbose,
    patience=patience
)
```

```python
from keras.optimizers import Adam
import tensorflow as tf
# Compile model sebelum training
model.compile(
    optimizer= Adam(learning_rate=adam),  # Use the defined `adam` variable
    loss='binary_crossentropy',
    metrics=['accuracy']
)
# Konversi generator menjadi tf.data.Dataset
train_dataset = tf.data.Dataset.from_generator(
    lambda: train_generator,
    output_signature=(
        tf.TensorSpec(shape=(None, *train_generator.image_shape), dtype=tf.float64),
        tf.TensorSpec(shape=(None,), dtype=tf.float64)
    )
).repeat()

validation_dataset = tf.data.Dataset.from_generator(
    lambda: validation_generator,
    output_signature=(
        tf.TensorSpec(shape=(None, *validation_generator.image_shape), dtype=tf.float64),
        tf.TensorSpec(shape=(None,), dtype=tf.float64)
    )
).repeat()

%time
hist = model.fit(
    train_dataset,
    validation_data=validation_dataset,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=epochs,
    callbacks=[lrd, mcp, es]

)
```

```python
import matplotlib.pyplot as plt
acc = hist.history['accuracy']
val_acc = hist.history['val_accuracy']
loss = hist.history['loss']
val_loss = hist.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'g', label='Training accuracy')
plt.plot(epochs, val_acc, 'y', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend(loc=0)
plt.figure()
plt.show()
```
